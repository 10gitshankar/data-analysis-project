```python
# 'h' bhanda agadi ko digit nikaalne
df['Duration_hour'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
# 'm' bhanda agadi ko digit nikaalne
df['Duration_min'] = df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)

#  (\d+): Yesle "digits" (numbers) lai matrai capture garcha.but error aunxa sakxa escape sequence error so use r'(\d+)h') 
#  h or m: Yesle tyo number ko pachadi 'h' ki 'm' chha bhanera check garcha.
#  .fillna(0): Yadi kunai flight 5h matrai chha bhane, tesma minute hudaina. Yo code le tyo khali thau ma 0 rakhidinchha.
```
